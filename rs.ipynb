```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
similarity = pd.DataFrame(cosine_similarity(np.array(vec).reshape(1,-1),pivot_table.to_numpy()).T, index=pivot_table.index, columns=['Cosine Similarity'])
```

```python
similarity.nlargest(5,'Cosine Similarity')
```

```python
avg_wage = df.drop(['league_name'],axis = 1).groupby('club_name').mean()['wage_eur']
```

```python
avg_wage[similarity.nlargest(5,'Cosine Similarity').index]
```

```python
def recommend(vec,k,pivot_table,avg_wage):
    np_vec = np.array(vec).reshape(1,-1)
    cs = cosine_similarity(np_vec,pivot_table.to_numpy())
    pd_cs = pd.DataFrame(cs.T,index=pivot_table.index,columns=['cosine_similarity'])
    ind = pd_cs.nlargest(k,'cosine_similarity').index
    wage = avg_wage[ind]
    return pd.DataFrame({'team_name':ind,'wage_eur':wage}).reset_index(drop=True)
```

```python
recommend(vec,5,pivot_table,avg_wage)
```
